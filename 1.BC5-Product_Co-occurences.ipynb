```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
import networkx as nx
```

```python
retail_df=pd.read_csv("df_2019.csv")
```

```python
retail_df.head()
```

```python
retail_df.info()
```

```python
df=retail_df[["ProductFamily_ID","Point-of-Sale_ID","Date"]]
df.head()
```

```python
df_POS200=df[df["Point-of-Sale_ID"]==200]
df_POS200
```

```python
df_POS200['month'] = pd.DatetimeIndex(df_POS200['Date']).month
df_POS200.head()
```

```python
df_1=df_POS200[(df_POS200["month"]==1)|(df_POS200["month"]==2)|(df_POS200["month"]==3)]
df_2=df_POS200[(df_POS200["month"]==4)|(df_POS200["month"]==5)|(df_POS200["month"]==6)]
df_3=df_POS200[(df_POS200["month"]==7)|(df_POS200["month"]==8)|(df_POS200["month"]==9)]
df_4=df_POS200[(df_POS200["month"]==10)|(df_POS200["month"]==11)|(df_POS200["month"]==12)]
```

```python
df_1
```

```python
df_1=df_1[["ProductFamily_ID","Date"]]
df_2=df_2[["ProductFamily_ID","Date"]]
df_3=df_3[["ProductFamily_ID","Date"]]
df_4=df_4[["ProductFamily_ID","Date"]]
```

#### 1st quarter

```python
mba_1 = pd.pivot_table(df_1, index='Date', columns='ProductFamily_ID', 
                    aggfunc=lambda x: 1 if len(x)>0 else 0).fillna(0)
mba_1
```

```python
freq_itemsets = apriori(mba_1, min_support=0.20, use_colnames=True)
freq_itemsets.sort_values(by="support", ascending=False)
```

```python
rulesConfidence = association_rules(freq_itemsets, metric="confidence", min_threshold=0.2)
rulesConfidence.sort_values(by='confidence', ascending=False, inplace=True)
rulesConfidence
```

```python
# High Confidence and high Lift - complementary products
rulesConfidence_cp=rulesConfidence[(rulesConfidence['confidence'] >= 0.7) & (rulesConfidence['lift'] > 1)]
rulesConfidence_cp.sort_values(by=['confidence','lift'], ascending=False, inplace=True)
rulesConfidence_cp
```

```python
# Substitue products 
rulesLift = association_rules(freq_itemsets, metric="lift", min_threshold=0.0)
rulesLift.sort_values(by='lift', ascending=True, inplace=True)
rulesLift[(rulesLift['lift'] < 1.0)]
```
